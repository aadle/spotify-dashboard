```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)

data = np.random.uniform(0, 1, size=(52, 24))
x = np.linspace(0, 1, data.shape[-1])
X = np.arange(0, 24)
G = 1.5 * np.exp(-4 * x ** 2 ) + 1
print(G.shape, x.shape, X.shape, data.shape)
```

```python
fig, ax = plt.subplots(figsize=(4, 7), frameon=False, facecolor="black", dpi=100)

for idx in range(data.shape[0]):
    lw = 1.5 - idx/100
    xscale = 4 - idx/200
    ax.plot(x*xscale, idx+data[idx, :], color="black", lw=lw)
ax.set_ylim(-1, 60)
plt.show()
```
