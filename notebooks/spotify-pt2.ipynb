```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import json
```

```python
with open("../secrets/data-retriever.json") as f:
    client_json = json.load(f)
```

```python
sp = spotipy.Spotify(
    auth_manager=SpotifyOAuth(
        client_id=client_json["client_id"],
        client_secret=client_json["client_secret"],
        redirect_uri="http://127.0.0.1:8000/callback",
        scope="user-read-recently-played"
        # scope="user-library-read"
        )
    )
```

```python
from datetime import timezone, datetime
start = int( datetime(2025, 12, 1).timestamp())
end = int(datetime(2025, 12, 22).timestamp() * 1000)

recently_played = sp.current_user_recently_played(limit=50, after=start)
```
